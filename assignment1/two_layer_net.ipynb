You can make sure that the implementations are correct by running the following:

- **总损失（忽略正则项）**：$L=\dfrac{1}{N}\sum\limits_{i=1}^{N}L_i$
- **SVM Loss**:
  - 损失公式:
  $$
  \color{red}L_i = \sum\limits_{j\ne y_i} \max\left(0, s_j-s_{y_i}+1\right)
  $$
  - $score$ 的梯度：
  $$
  \nabla_{score_{y_i}} L_i = -\left(\sum\limits_{j\ne y_i} \mathbb{I}\left(s_j-s_{y_i}+\Delta >0 \right) \right), \quad \nabla_{score_j} L_i = \left(\sum\limits_{j\ne y_i} \mathbb{I}\left(s_j-s_{y_i}+\Delta >0 \right) \right)
  $$

- **Softmax Loss**:
  - 损失公式:
  $$
  \color{red}L_i = -\log\left(\dfrac{e^{f_{y_i}}}{\sum\limits_j e^{f_j}}\right) = -f_{y_i} + \log\left(\sum\limits_j e^{f_j}\right)
  $$
  - $f$ 的梯度：
  $$
  \nabla_{f_{y_i}} L_i = -1 + \dfrac{e^{f_{y_i}}}{\sum\limits_j e^{f_j}}, \quad \nabla_{f_j} L_i = \dfrac{e^{f_j}}{\sum\limits_j e^{f_j}}
  $$
